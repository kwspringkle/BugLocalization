Tiền xử lý các Bug Reports của từng project

```python
# Tiền xử lý Bug Reports aspectj
aspectj_bug_reports = Parser(aspectj).report_parser()  # Đọc bug reports
aspectj_report_preprocessor = ReportPreprocessing(aspectj_bug_reports)
aspectj_report_preprocessor.preprocess()

# Tiền xử lý Bug Reports eclipse
eclipse_bug_reports = Parser(eclipse).report_parser()  # Đọc bug reports
eclipse_report_preprocessor = ReportPreprocessing(eclipse_bug_reports)
eclipse_report_preprocessor.preprocess()

# Tiền xử lý Bug Reports swt
swt_bug_reports = Parser(swt).report_parser()  # Đọc bug reports
swt_report_preprocessor = ReportPreprocessing(swt_bug_reports)
swt_report_preprocessor.preprocess()

# Tiền xử lý Bug Reports tomcat
tomcat_bug_reports = Parser(tomcat).report_parser()  # Đọc bug reports
tomcat_report_preprocessor = ReportPreprocessing(tomcat_bug_reports)
tomcat_report_preprocessor.preprocess()

# Tiền xử lý Bug Reports birt
birt_bug_reports = Parser(birt).report_parser()  # Đọc bug reports
birt_report_preprocessor = ReportPreprocessing(birt_bug_reports)
birt_report_preprocessor.preprocess()
```

Tiền xử lý cho các Source files của từng project

```python
# Tiền xử lý Source files aspectj
aspectj_src_reports = Parser(aspectj).src_parser()
aspectj_src_preprocessor = SrcPreprocessing(aspectj_src_reports)
aspectj_src_preprocessor.preprocess()

# Tiền xử lý Source files eclipse
eclipse_src_reports = Parser(eclipse).src_parser()
eclipse_src_preprocessor = SrcPreprocessing(eclipse_src_reports)
eclipse_src_preprocessor.preprocess()

# Tiền xử lý Source files swt
swt_src_reports = Parser(swt).src_parser()
swt_src_preprocessor = SrcPreprocessing(swt_src_reports)
swt_src_preprocessor.preprocess()

# Tiền xử lý Source files tomcat
tomcat_src_reports = Parser(tomcat).src_parser()
tomcat_src_preprocessor = SrcPreprocessing(tomcat_src_reports)
tomcat_src_preprocessor.preprocess()

# Tiền xử lý Source files birt
birt_src_reports = Parser(birt).src_parser()
birt_src_preprocessor = SrcPreprocessing(birt_src_reports)
birt_src_preprocessor.preprocess()
```
