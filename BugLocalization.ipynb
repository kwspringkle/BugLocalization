## 2.Trích xuất đặc trưng

### 2.1. Lexical similarity
Tính độ tương đồng mức từ giữa bug reports và source code  
Sử dụng TFIDF Vectorizer

### 2.2. Semantic similarity
Do có lexical mismatch khi bug reports sử dụng ngôn ngữ tự nhiên, còn source code lại sử dụng ngôn ngữ lập trình --> Tìm hiểu thêm về semantic relationship giữa các từ phụ thuộc vào context nó xuất hiện

- Sử dụng Glove để chuyển từ --> vectors

### 2.3. Similar Bug Reports  
Dựa trên các bug reports trước đó mà source file đã được fix

### 2.4. Code change history  
Các file mà đã fix xong rất lâu về trước hoặc chưa bao giờ fix thì khả năng sẽ ít mắc lỗi hơn

### 2.5. Bug fixing frequency  
Số lần mà source file đã được sửa cũng sẽ ảnh hưởng đến xác suất file đó lỗi hay không
